```python
import matplotlib.pyplot as plt
import seaborn as sns

# Separate numeric and categorical columns
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_cols = df.select_dtypes(exclude=['int64', 'float64']).columns.tolist()

sns.set(style="whitegrid")

# ---------- Numeric Columns ----------
for col in numeric_cols:
    fig, axes = plt.subplots(1, 2, figsize=(12, 4))

    # Histogram
    sns.histplot(df[col], bins=30, kde=True, ax=axes[0], color="skyblue")
    axes[0].set_title(f"Distribution of {col}")

    # Boxplot
    sns.boxplot(x=df[col], ax=axes[1], color="lightcoral")
    axes[1].set_title(f"Boxplot of {col}")

    plt.tight_layout()
    plt.show()

# ---------- Categorical Columns ----------
for col in categorical_cols:
    plt.figure(figsize=(10, 4))

    # Value counts bar plot
    df[col].value_counts().plot(kind='bar', color="lightgreen")
    plt.title(f"Counts of {col}")
    plt.ylabel("Frequency")
    plt.xlabel(col)
    plt.xticks(rotation=45)

    plt.show()
```

```python
# Correlation heatmap for all numeric features
plt.figure(figsize=(12, 8))
corr = df[numeric_cols].corr()
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm", cbar=True)
plt.title("Correlation Heatmap of Numeric Features")
plt.show()
```

```python
# Scatter plots of AQI vs all other numeric features
for col in numeric_cols:
    if col != "aqi":  # exclude AQI itself
        plt.figure(figsize=(6, 4))
        sns.scatterplot(x=df[col], y=df['aqi'], alpha=0.5)
        plt.title(f"AQI vs {col}")
        plt.xlabel(col)
        plt.ylabel("AQI")
        plt.show()
```

```python
# AQI distribution across districts
plt.figure(figsize=(12, 5))
sns.boxplot(x="district", y="aqi", data=df)
plt.title("AQI Distribution by District")
plt.xticks(rotation=45)
plt.show()

# AQI distribution across divisions
plt.figure(figsize=(10, 5))
sns.boxplot(x="division", y="aqi", data=df)
plt.title("AQI Distribution by Division")
plt.xticks(rotation=45)
plt.show()

# AQI trend over time (if 'date' is usable as datetime)
df['date'] = pd.to_datetime(df['date'], errors='coerce')  # ensure datetime format
plt.figure(figsize=(12, 5))
sns.lineplot(x="date", y="aqi", data=df, ci=None)
plt.title("AQI Trend Over Time")
plt.xticks(rotation=45)
plt.show()
```

```python
sns.pairplot(df[["aqi", "pm2_5", "pm10", "o3", "no2", "so2", "co"]], diag_kind="kde")
plt.suptitle("Pairplot of AQI and Pollutants", y=1.02)
plt.show()
```

```python
# Average AQI by division and district
pivot_table = df.pivot_table(values="aqi", index="division", columns="district", aggfunc="mean")

plt.figure(figsize=(12, 6))
sns.heatmap(pivot_table, annot=True, fmt=".1f", cmap="YlOrRd")
plt.title("Average AQI by Division and District")
plt.show()
```

```python
# Extract month & season
df["month"] = df["date"].dt.month
df["season"] = pd.cut(df["month"],
                      bins=[0, 3, 6, 9, 12],
                      labels=["Winter", "Spring", "Summer", "Autumn"])

plt.figure(figsize=(12, 5))
sns.boxplot(x="season", y="aqi", hue="division", data=df)
plt.title("AQI Distribution by Season and Division")
plt.show()
```
